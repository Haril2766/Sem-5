### 24) histogram for Age attribute

```python
import matplotlib.pyplot as plt
f['age'].dropna().hist(bins=8,edgecolor='black')
plt.title("Histogram of age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### 25) A boxplot can also be used to show the distribution of values for each attribute.

```python
numeric = ['age', 'trestbps']
data= f[numeric].dropna()

plt.figure(figsize = (4,4))
data.boxplot(column = numeric)
plt.title("Boxplot of Heart Numeric Attributes")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```
